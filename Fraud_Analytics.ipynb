```python
y = Data['isFraud']
time = Data['step']
```

```python
Data = Data.drop(columns = ['isFraud','nameOrig','nameDest'])
```

```python
Data
```

```python
def code(x):
    if(x=='TRANSFER'):
        return 0
    elif(x=='CASH_OUT'):
        return 1
    elif(x=='CASH_IN'):
        return 2
    elif(x=='PAYMENT'):
        return 3
    elif(x=='DEBIT'):
        return 4
Data['Coded'] = Data.type.transform(code)
```

```python
Data = Data.drop(columns= 'type')
```

```python
from imblearn.over_sampling import SMOTE

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Data, y, test_size=0.3, random_state=0)
```

```python
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_train, y_train)
```

```python
X_train_res = pd.DataFrame(X_train_res)
```

```python
Data_t = pd.DataFrame()
X_train_res
```

```python
Data_t['isFraud'] = y_train_res
Data_t = Data_t.drop(columns = [0,1,2,3,4,5,6,7])
```

```python
def trans(x):
    if(x == 0):
        return 'TRANSFER'
    elif(x==1):
        return 'CASH_OUT'
    elif(x==2):
        return 'CASH_IN'
    elif(x==3):
        return 'PAYMENT'
    elif(x==4):
        return 'DEBIT'
Data_t['type'] = Data_t.Coded.transform(trans)
```

```python
def applymode(x):
    a = x.mode()[0]
    print(a)
    return x.fillna(a)
```

```python
from bokeh.core.properties import value
from bokeh.io import show, output_file,output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
output_notebook

output_file("bar_stacked_grouped.html")

factors = ["TRANSFER","CASH_OUT","CASH_IN","PAYMENT","DEBIT"]

regions = ['Zero', 'One']

source = ColumnDataSource(data=dict(
    x=factors,
    Zero=[ 5458,23426,14584,22810, 431 ],
    One=[ 39570,27139,0,0,0 ],
))

q = figure(x_range=FactorRange(*factors), plot_height=250,
           toolbar_location=None, tools="")

q.vbar_stack(regions, x='x', width=0.9, alpha=0.5, color=["blue", "red"], source=source,
             legend=[value(x) for x in regions])

q.y_range.start = 0
q.y_range.end = 55000
q.x_range.range_padding = 0.1
q.xaxis.major_label_orientation = 1
q.xgrid.grid_line_color = None
q.legend.location = "top_center"
q.legend.orientation = "horizontal"

show(q)
```
